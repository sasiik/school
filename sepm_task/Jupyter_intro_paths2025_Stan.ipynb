## Plotting quantitative data
- Histogram is used to show the distribution of quantitative data
- plot.hist using Matplotlib
- sns.histplot using Seaborn

### Histogram with Matplotlib
Read examples of creation of histogram:
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
- https://github.com/ageron/handson-ml3/blob/main/tools_matplotlib.ipynb

```python
# Help
plt.hist
```

```python
plt.hist(trips.Age, bins=10, color='r',edgecolor='blue')   
plt.xlabel('Age', fontsize=16)
plt.ylabel("Freq")
plt.title("Histogram of travellers' age", fontsize=18)
plt.show() 
```

### Histogram with Seaborn
- Seaborn is another library for graphics
- Based on Matplotlib
- histplot is a function in Seaborn for Historgrams 
- https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
sns.histplot(data = trips, x= 'Age', color='yellow', 
             edgecolor ='red',bins=10, hue = 'Gender') 
plt.xlabel('Age', fontsize=14)
plt.ylabel('Freq', fontsize=12)   
plt.title("Histogram of travellers' age per Gender", fontsize=18)
plt.show()
```

## DataFrame is created using CSV file

```python
df1 = pd.read_csv('Cycling.csv', decimal = ',')
df1.head()
```

### Task 3 
- Explore the data
- use describe, make histogram, bar graphs

```python
# your code in here
from matplotlib.ticker import MaxNLocator

df1.describe()
keys = ['Distance', 'Avg HR', 'Max HR', 'Aerobic TE', 'Min Temp', 'Best Lap Time', 'Max Temp', 'Time', 'Moving Time', 'Elapsed Time']
for key in keys:
    plt.hist(sorted(df1[key]), bins=df1.index.size, color='r',edgecolor='blue')   
    plt.ylabel("Count", fontsize=16)
    plt.xlabel("Value")
    ax = plt.gca()
    ax.xaxis.set_major_locator(MaxNLocator(nbins=5)) 
    plt.title(f"Histogram of travellers' {key}", fontsize=18)
    plt.show() 
```
